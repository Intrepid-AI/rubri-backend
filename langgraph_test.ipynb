````python
import os
import json
from typing import Annotated, Dict, List, Literal, Optional, Sequence, TypedDict, Union, Any
from pydantic import BaseModel, Field
from langchain_core.messages import AIMessage, BaseMessage, HumanMessage
from langchain_openai import ChatOpenAI  # Changed to OpenAI
from langchain_core.tools import tool
from langgraph.graph import StateGraph, END
from langgraph.graph.message import add_messages
from langgraph.checkpoint.memory import MemorySaver
from langgraph.prebuilt import create_react_agent
from langgraph.types import Command
from langgraph.graph import StateGraph, END, START
````

````python
## 2. Define Data Models for Input and Output

class ScoringCriteria(BaseModel):
    """Scoring criteria for a skill assessment."""
    level_1: str = Field(..., description="Poor performance description")
    level_2: str = Field(..., description="Below average performance description")
    level_3: str = Field(..., description="Average performance description")
    level_4: str = Field(..., description="Above average performance description")
    level_5: str = Field(..., description="Excellent performance description")

class RubricItem(BaseModel):
    """An individual item in the rubric."""
    skill: str = Field(..., description="Name of the skill or competency")
    description: str = Field(..., description="Description of the skill and why it's important")
    scoring_criteria: ScoringCriteria = Field(..., description="Criteria for scoring this skill")
    sample_questions: List[str] = Field(..., description="Sample interview questions to assess this skill")

class RubricSection(BaseModel):
    """A section in the rubric containing related items."""
    name: str = Field(..., description="Name of the section (e.g., 'Technical Skills')")
    description: Optional[str] = Field(default=None, description="Description of this section")
    items: List[RubricItem] = Field(..., description="Skills or competencies in this section")

class InterviewRubric(BaseModel):
    """The complete interview rubric."""
    title: str = Field(..., description="Title of the rubric")
    position_summary: str = Field(..., description="Brief summary of the position")
    sections: List[RubricSection] = Field(..., description="Sections of the rubric")
    general_interview_guidance: Optional[str] = Field(default=None, description="General guidance for interviewers")
````

````python
## 3. Define the Agent State

class RubricGeneratorState(TypedDict):
    """The state maintained throughout the rubric generation process."""
    # Conversation history
    messages: Annotated[Sequence[BaseMessage], add_messages]
    
    # Processing state
    resume_text: Optional[str]
    job_description: Optional[str]
    position_title: Optional[str]
    
    # Output data
    rubric: Optional[Dict[str, Any]]
````

````python
## 7. Define Node Functions for the StateGraph

def initialize_state(state: RubricGeneratorState) -> RubricGeneratorState:
    """Initialize the state with defaults."""
    # Examine the last message to extract parameters
    if len(state["messages"]) > 0 and isinstance(state["messages"][-1], HumanMessage):
        content = state["messages"][-1].content
        
        # We're keeping this simple for the experiment - in a real implementation,
        # we would do more sophisticated parsing of the input
        
        # Check if there's a resume in the message
        if "RESUME:" in content:
            parts = content.split("RESUME:")
            if len(parts) > 1:
                resume_part = parts[1]
                if "JOB DESCRIPTION:" in resume_part:
                    state["resume_text"] = resume_part.split("JOB DESCRIPTION:")[0].strip()
                elif "POSITION TITLE:" in resume_part:
                    state["resume_text"] = resume_part.split("POSITION TITLE:")[0].strip()
                else:
                    state["resume_text"] = resume_part.strip()
        
        # Check if there's a job description in the message
        if "JOB DESCRIPTION:" in content:
            parts = content.split("JOB DESCRIPTION:")
            if len(parts) > 1:
                jd_part = parts[1]
                if "POSITION TITLE:" in jd_part:
                    state["job_description"] = jd_part.split("POSITION TITLE:")[0].strip()
                else:
                    state["job_description"] = jd_part.strip()
        
        # Check if there's a position title in the message
        if "POSITION TITLE:" in content:
            parts = content.split("POSITION TITLE:")
            if len(parts) > 1:
                state["position_title"] = parts[1].strip()
        
    return state
````

````python
def analyze_inputs(state: RubricGeneratorState) -> RubricGeneratorState:
    """Analyze resume and job description to extract skills and requirements."""
    # In a real implementation, we would call the tools to extract skills
    # For this experiment, we'll simulate that step
    
    # Add a message to the state to record this step
    state["messages"].append(AIMessage(
        content="I've analyzed the provided resume and job description. Now I'll generate a comprehensive interview rubric based on these inputs."
    ))
    
    return state
````

````python
def generate_rubric(state: RubricGeneratorState) -> RubricGeneratorState:
    """Generate the final rubric structure."""
    # Prepare prompt for the LLM
    resume_text = state.get("resume_text", "")
    job_description = state.get("job_description", "")
    position_title = state.get("position_title", "Interview Candidate")
    
    prompt =  f"""
    Based on the following information, create a detailed TECHNICAL interview rubric:
    
    Position Title: {position_title}
    """
    
    if resume_text:
        prompt += f"\n\nRESUME:\n{resume_text}"
    
    if job_description:
        prompt += f"\n\nJOB DESCRIPTION:\n{job_description}"
    
    prompt += """
    
    Create a comprehensive TECHNICAL interview rubric specifically for technical rounds that includes:
    1. Multiple relevant sections (at least 3)
    2. At least 3 skills per section
    3. Detailed scoring criteria for each skill (levels 1-5)
    4. 2-3 HIGHLY TECHNICAL interview questions per skill that test deep conceptual understanding
    
    REMEMBER: Questions must be deeply technical, testing core concepts and implementation details, not just general experience.
    
    Format the output as a structured JSON object matching the InterviewRubric schema.
    """
    
    # Call the LLM directly for this step
    llm_response = model.invoke(prompt)
    
    # Extract JSON from the LLM response
    response_text = llm_response.content
    
    # Try to find and parse JSON in the response
    try:
        # Look for JSON block in markdown
        if "```json" in response_text:
            json_text = response_text.split("```json")[1].split("```")[0].strip()
        elif "```" in response_text:
            json_text = response_text.split("```")[1].split("```")[0].strip()
        else:
            json_text = response_text
            
        # Parse the JSON
        rubric_data = json.loads(json_text)
        state["rubric"] = rubric_data
        
    except Exception as e:
        # If JSON parsing fails, add an error message
        state["messages"].append(AIMessage(
            content=f"I encountered an error while generating the rubric structure: {str(e)}\n\nHere's the raw response:\n{response_text}"
        ))
        return state
    
     # Add the final response to messages
    state["messages"].append(AIMessage(
        content=f"I've created an technical interview rubric for the {position_title} position. The rubric includes multiple sections with detailed scoring criteria and highly technical questions that test deep conceptual understanding."
    ))
    
    return state
````

````python
def should_continue(state: RubricGeneratorState) -> Literal["initialize", "analyze", "generate", "complete"]:
    """Determine the next step in the workflow."""
    # If we haven't extracted input parameters yet
    if not state.get("resume_text") and not state.get("job_description") and not state.get("position_title"):
        return "initialize"
    
    # If we have inputs but haven't analyzed them yet
    if not any(msg.content.startswith("I've analyzed") for msg in state["messages"] if isinstance(msg, AIMessage)):
        return "analyze"
    
    # If we've analyzed inputs but haven't generated the rubric yet
    if not state.get("rubric"):
        return "generate"
    
    # If we've generated the rubric, we're done
    return "complete"
````

````python
# Create the memory checkpoint saver
memory = MemorySaver()

# Create the graph
workflow = StateGraph(RubricGeneratorState)

# Add nodes
workflow.add_node("initialize", initialize_state)
workflow.add_node("analyze", analyze_inputs)
workflow.add_node("generate", generate_rubric)

# Set entry point - this is critical!
workflow.add_edge(START, "initialize")

# Add conditional edges
workflow.add_conditional_edges(
    "initialize",
    should_continue,
    {
        "initialize": "initialize",  # Loop back if we need more processing
        "analyze": "analyze",
        "generate": "generate",
        "complete": END
    }
)

workflow.add_conditional_edges(
    "analyze",
    should_continue,
    {
        "initialize": "initialize",
        "analyze": "analyze",  # Loop back if needed
        "generate": "generate",
        "complete": END
    }
)

workflow.add_conditional_edges(
    "generate",
    should_continue,
    {
        "initialize": "initialize",
        "analyze": "analyze",
        "generate": "generate",  # Loop back if needed
        "complete": END
    }
)

# Compile the graph WITH the checkpointer
rubric_generator = workflow.compile(checkpointer=memory)
````

````python
## 10. Function to Process User Input

def generate_interview_rubric(
    position_title: str, 
    resume_text: Optional[str] = None, 
    job_description: Optional[str] = None,
    thread_id: str = "default"
) -> Dict:
    """
    Generate a technical interview rubric based on the provided information.
    
    Args:
        position_title: The title of the position
        resume_text: Optional text of the candidate's resume
        job_description: Optional text of the job description
        thread_id: Unique identifier for the conversation thread
        
    Returns:
        A dictionary containing the generated rubric and conversation history
    """
    # Prepare the input message
    message_content = f"POSITION TITLE: {position_title}\n\n"
    
    if resume_text:
        message_content += f"RESUME: {resume_text}\n\n"
        
    if job_description:
        message_content += f"JOB DESCRIPTION: {job_description}\n\n"
    
    messages = [HumanMessage(content=message_content)]
    
    # Run the agent
    config = {"configurable": {"thread_id": thread_id}}
    result = rubric_generator.invoke(
        {
            "messages": messages,
            "resume_text": None,
            "job_description": None,
            "position_title": None,
            "rubric": None
        },
        config
    )
    
    return result

````
